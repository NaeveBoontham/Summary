# Merge (JOIN of SQL)

---

```python
movie_dict = {"movie_id": [101101,101102,101103,101104,101105,101106,101107,101108,101109,101110],
                     "movie": ["Django", "Drive", "Spiderman", "Superman", "GoodFellas", "The Host", "Interstella", "Inception", "Blood Diamond", "Conjuring"],
                     "release_year": ["2012", "2011", "2012", "2013", "1990", "2006", "2014", "2010", "2006", "2013"]}

actor_dict = {"id": [101,102,103,104,105,106,107,108,109,110],
                    "actor": ["Jamie Fox", "Ryan Gosling", "Andrew Garfield", "Henry Cavil", "Robert De Niro", "Song Kang Ho", "Matthew Mcconaughey",
                                 "Leonado Dicaprio", "Leonardo Dicaprio", "Patrick Wilson"]}

gross_dict = {"movie_id": [101101,101102,101103,101104,101105,101106,101107,101108,101109,101110],
                     "actor_id": [101,102,103,104,105,106,107,108,109,110],
                     "gross_domestic(million)": [240, 120, 410, 360, 200, 110, 500, 440, 220, 300],
                     "gross_foreign(million)": [120, 60, 205, 180, 100, 55, 250, 220, 110, 150],
                     "score": [89,80,85,78,92,85,90,88,75,67]}

movie = pd.DataFrame(movie_dict)
actor = pd.DataFrame(actor_dict)
gross = pd.DataFrame(gross_dict)
```

```python
movie
```

```python
actor
```

```python
gross
```

## pd.merge()

#### Normal merge (merge column have the same name)

```python
# Use merge to join gross table and movie table 
# Get ouput as movie name, release year, domestice gross, foreign gross, score

pd.merge(gross, movie, on = "movie_id", how = "inner")[["movie", "release_year", "gross_domestic(million)", "gross_foreign(million)", "score"]]
```

```python
# Use merge to join gross table and movie table 
# Get ouput as movie name, release year, domestice gross, foreign gross, score
# OTHER WAY

gross.merge(movie, on = "movie_id", how = "inner")[["movie", "release_year", "gross_domestic(million)", "gross_foreign(million)", "score"]]
```

#### left_on,right_on (if column dont have the same name)

```python
# Use merge to join gross table and actor table
# Get output as actor name, domestic gross, foreign gross, score

pd.merge(gross, actor, left_on = "actor_id", right_on = "id", how = "inner")[["actor", "gross_domestic(million)", "gross_foreign(million)", "score"]]
```

#### left_index, right_index (merge using index)

```python
# Use merge to join gross table and movie table using index
# Get ouput as movie name, release year, domestice gross, foreign gross, score

pd.merge(gross, movie, left_index = True, right_index = True, how = "inner")[["movie", "release_year", "gross_domestic(million)", "gross_foreign(million)", "score"]]
```

# Group by AND Aggregate Function (Group by of SQL)

---

```python
university_dict = {"student": ["George", "Isora", "Pluem", "Aum", "Luis", "Jui", "Sam", "Barnes", "June", "Jame", "Non", "Mark", "Muichi", "Henry", "Kram", "Mew", "Kane", "Robert", "Hill", "Jenna"],
                          "department": ["Electrical", "Civil", "Mechanical", "Chemical", "Computer", "Electrical", "Civil", "Mechanical", "Chemical", "Computer", "Electrical", "Civil", "Mechanical", "Chemical", "Computer",
                                                "Electrical", "Civil", "Mechanical", "Chemical", "Computer"],
                          "country": ["Thailand", "USA", "China", "Brazil", "Spain", "Spain", "China", "Thailand", "Brazil", "USA", "China", "Spain", "Brazil", "USA", "Thailand", "Brazil", "Spain", "Thailand", "USA", "China"],
                          "grade": ["A", "A", "C", "B", "C", "C", "B", "A", "D", "B", "C", "A", "A", "C", "B", "B", "C", "A", "D", "F"],
                          "score": [82,90,66,80,63,69,76,92,57,73,61,95,86,60,79,77,65,100,52,13]}

university = pd.DataFrame(university_dict)
```

```python
university
```

```python
# Group university by department

university_group = university.groupby("department")
```

```python
# Use dir() to see all attribute and method of group
dir(university_group)
```

#### get_group()

```python
# Use get_group to access the group you want

university_group.get_group("Computer")
```

```python
university_group.get_group("Electrical")
```

#### max(), min(), mean(), median()

```python
university_group.max("score")
```

```python
university_group.min("score")
```

```python
university_group.mean("score")
```

```python
university_group.median("score")
```

## Group and Aggregate

### .agg() Per column (Use dictionary)

```python
# Use as_index = False 
university.groupby(by = "department", as_index = False).agg({"score": ["sum", "min", "max", "mean"]})
```

```python
# Use .reset_index()
university.groupby(by = "department").agg({"score": ["sum", "min", "max", "mean"]}).reset_index()
```

```python
# If you not use as_index = False OR reset_index() the column that use to group by will become the index itself
university.groupby(by = "department").agg({"score": ["sum", "min", "max", "mean"]})
```

### .agg() Over new column (Create new column to do aggregate)

```python
# Create new column name score_min, score_max, score_mean and aggregate over it
university.groupby(by = "department").agg(score_min = ("score", "min"), score_max = ("score", "max"), score_mean = ("score", "mean"))
```

### Group more than 1 column

```python
# Groupby department, grade
university.groupby(by = ["department", "grade"]).agg(score_min = ("score", "min"), score_max = ("score", "max"), score_mean = ("score", "mean"))
```

```python
# Groupby country, grade
university.groupby(by = ["country", "grade"]).agg(grade_count = ("grade", "count"), score_min = ("score", "min"), score_max = ("score", "max"), score_mean = ("score", "mean"))
```

```python
# Groupby department, country, grade
university.groupby(by = ["department", "country", "grade"]).agg(grade_count = ("grade", "count"), score_min = ("score", "min"), score_max = ("score", "max"), score_mean = ("score", "mean"))
```
